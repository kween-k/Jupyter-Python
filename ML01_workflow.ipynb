```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
URL = pd.read_json('https://data.seattle.gov/resource/4xy5-26gy.json')
```

```python
df = pd.DataFrame(URL)
df.set_index('date', inplace=True)
print(df.head())
```

```python
df.plot()
plt.show()
```

```python
#To see a little better, we can resample using sum
df.resample('W').sum().plot();
```
